## ANOVA란 집단 내 자료의 변동을 고려했을 때, 집단 간 평균의 차이가 유의미하게 다른지 통계적으로 검정하는 것입니다.
### 귀무 가설은 각 요일 별 활동량의 평균이 같다는 것이고, 대립 가설은 그 중 하나 이상이 다른 것입니다. 
### ANOVA를 하기 위해선 정규성, 등분산성, 독립성을 만족해야 합니다. 
### 정규성은 Shapiro Test, 독립성은 Durbin-Watson Test, 등분산성은 Bartlett Test를 사용합니다.

### 만일 저 가정을 만족하지 않는 데이터셋이라면, 이상치를 제거하고 ANOVA를 시행하면 좋습니다.

## 검정 통계량 F에 대해 p-value가 기준치 아래 (0.05)라면 귀무 가설이 기각되었다고 할 수 있습니다.

```python
# DataFrame으로 저장합니다
df = pd.DataFrame({'log_date':log_date, 'push_count':push_count})
df.log_date = pd.to_datetime(df.log_date, format='%y-%m-%d')
df = df.assign(day_of_week = df.log_date.dt.dayofweek)
```

```python
# 구간을 설정합니다.
x_max = df.push_count.max()
x_min = df.push_count.min()
bins = np.arange(x_min, x_max, step=(x_max-x_min)/20)
```

```python
# 요일별로 나누어 시각화합니다.
plt.figure(figsize=(10,5))
legs = []
mean_dict = {}
for dow, group in df.groupby('day_of_week'):
    group.push_count.hist(bins=bins, alpha=0.3)
    mean_dict[dow] = group.push_count.mean()
    legs.append(dow)


plt.legend(legs)
plt.ylabel('Day count')
plt.xlabel('Push count')
plt.show()
```

```python
# 요일별 평균을 mean_dow에 저장합니다.
# 평균과의 차이를 residual로 저장하고, 정규화합니다.
df = df.assign(mean_dow=df['day_of_week'].map(mean_dict))
df['residual'] = df['push_count'] - df['mean_dow']

# df['residual'].mean()은 0에 매우 가깝습니다.
print(df['residual'].mean())

df['std_resi'] = (df['residual'] - df['residual'].mean())/df['residual'].std()
# 이동평균치를 push_count_moving_average로 저장합니다.
df = df.assign(push_count_ma = df.push_count.rolling(window=28).mean())

df.tail(56)
```

```python
# MA plot - Nan인 부분은 자동으로 제외됩니다
# 연말에 push가 적다는 것을 볼 수 있습니다.
plt.plot(df.log_date, df.push_count_ma)
plt.xticks(rotation=20)
plt.show()
```

```python
# QQ plot
# ANOVA의 3가지 가정 중 하나인 Normality를 확인하기 위해 QQ plot을 진행합니다.
# 정규분포와 가깝지 않다는 것을 알 수 있습니다.
sm.qqplot(df['std_resi'], line='45')
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Standardized Residuals")
plt.show()
```

```python
categorized = []
for dow, group in df.groupby('day_of_week'):
    categorized.append(group.push_count)
```

```python
# normality
w, pvalue = stats.shapiro(df['residual'])
print(f"Normaility test p-value: {pvalue:.4f}")

# equal variance
w, pvalue = stats.bartlett(*categorized)
print(f"Equal variance test p-value: {pvalue:.4f}")

# independency
d = durbin_watson(df['residual'])
print(f"Independence test statistic: {d:.4f}")
```

## p-value는 가정한 분포에서 지금의 상황보다 더욱 극단적인 상황이 일어날 확률을 의미합니다. 
## 따라서 p-value가 매우 낮은 것은 이미 현 상황이 극단적인(전제가 틀린) 상황이라는 뜻이고, 
## 정규분포가 **아님**을 의미합니다. 이제 이상치를 제거해보겠습니다

```python
def z_score_outlier_remover(l, dow, threshold):
    total_outlier_count = 0
    l = np.array(l)
    
    while True:
        m = np.mean(l)
        s = np.std(l, ddof=1)
        outlier_bool = (abs(l-m)/s>threshold)
        outlier_count = outlier_bool.sum()
        if outlier_count > 0:
            total_outlier_count += outlier_count
            l = [i for i, bool in zip(l, outlier_bool) if not bool]
        else:
            if total_outlier_count > 0:
                print(f"The number of z-score outliers on {dow}: {total_outlier_count}")
                break
    return l

df_no_outlier = pd.concat([pd.DataFrame({'push_count':z_score_outlier_remover(group.push_count, dow, 2.5), 'day_of_week':dow}) for dow, group in df.groupby('day_of_week')])

categorized = []
mean_dict = {}
for dow, group in df_no_outlier.groupby('day_of_week'):
    mean_dict[dow] = group.push_count.mean()
    categorized.append(group.push_count)  

df_no_outlier = df_no_outlier.assign(mean_dow=df_no_outlier['day_of_week'].map(mean_dict))
df_no_outlier['residual'] = df_no_outlier['push_count'] - df_no_outlier['mean_dow']
```

```python
# normality
w, pvalue = stats.shapiro(df_no_outlier['residual'])
print(f"Normaility test p-value: {pvalue:.4f}")

# equal variance
w, pvalue = stats.bartlett(*categorized)
print(f"Equal variance test p-value: {pvalue:.4f}")

# independency
d = durbin_watson(df_no_outlier['residual'])
print(f"Independence test statistic: {d:.4f}")

# QQ plot
df_no_outlier['std_resi'] = (df_no_outlier['residual'] - df_no_outlier['residual'].mean())/df_no_outlier['residual'].std()
sm.qqplot(df_no_outlier['std_resi'], line='45')
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Standardized Residuals")
plt.show()
```

이상치를 제거하니 p-value가 0.05이상으로 올라가 덜 극단적인 상황이 되었습니다.

```python
# ANOVA
F, p_value = stats.f_oneway(*categorized)
print(f"F statistics: {F:.4f}")
print(f"P-value: {p_value:.4f}")
```
